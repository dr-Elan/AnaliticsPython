```python
df = pd.read_csv('https://gbcdn.mrgcdn.ru/uploads/asset/4957768/attachment/79dce5d9a8cff55bf57542ae9b63338f.csv', encoding='windows-1251')
```

```python
df['Cpu_company'] = df['Cpu'].apply(lambda x: x.split()[0])
df
```

```python
some_string = '128GB SSD'
some_string.split('GB')[0]
```

```python
df['Memory_Amount'] = df['Memory'].apply(lambda x: x.split('GB')[0])
df.loc[:,'Memory_Amount'] = df['Memory_Amount'].apply(lambda x: float((x.split('TB')[0])) )
df
```

```python
df.drop(['Memory', 'ScreenResolution'], axis = 1, inplace= True)
```

```python
df
```

```python
%%time
def getb(my_string):
    keywords = {"HDD", "SSD", "Flash", "Hybrid"}
    myset=set(my_string.split()) & set(keywords)
    return ', '.join(myset)

df['Memory_Type'] = df['Memory'].apply(getb)
df
```
